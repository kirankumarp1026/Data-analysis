```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Download the Boston housing dataset from Kaggle (replace this)
# Replace 'your_downloaded_file.csv' with the actual filename
data = pd.read_csv('/content/drive/MyDrive/Boston.csv')
# Select features and target
features = ['crim', 'rm']  # Example: Select two features initially
target = 'medv'

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data[features], data[target], test_size=0.2, random_state=42)

# Create the linear regression model
model = LinearRegression()
# Fit the model to the training data
model.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = model.predict(X_test)
```

```python
# Plot the results for CRIM vs. MEDV
plt.figure(figsize=(8, 6))
plt.scatter(X_test['crim'], y_test, color='blue', label='Actual Price')
plt.plot(X_test['crim'], y_pred, color='red', label='Predicted Price')
plt.xlabel('Crime Rate (per capita)')
plt.ylabel('Median Value of Homes')
plt.title('Linear Regression: crim vs. medv')
plt.legend()
plt.grid(True)
plt.show()
```
